```python
class employees():
    
    def __init__(self,name,age,language):
        self.name = name
        self.age = age
        self.lang = language
        
    def employ(self):
        print(self.name, "is an employee and his age is",self.age,"and he can speak",self.lang)
        
employee1 = employees("David",45,"Engish")
employee1.employ()
```

```python
class managers(employees):
    pass
```

```python
class managers(employees):
    
    def manage(self):
        print(self.name,"is a manager and his/her age is",self.age,"and he/she can speak",self.lang)
        
manager4 = managers("Kerin",34,"Arabic")
manager4.manage()
```
