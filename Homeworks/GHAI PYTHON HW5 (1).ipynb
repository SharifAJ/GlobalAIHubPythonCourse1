```python
class Animals():
    def __init__(self,color,name,legs):
        self.name = name
        self.leg = legs
        self.eye = color
        
    def Animal(self):
        
        print("the",self.name,"has",self.leg,"and its eye color is",self.eye)
        

Tiger3 = Animals("Blue","Tiger","4 Legs")
print(Tiger3)
Tiger3.Animal()
```

```python
class Cats(Animals):
    pass

cat = Cats("Grey","Persian Cat","4 Legs")
cat.Animal()
```
